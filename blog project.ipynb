```python
def get_desc ( col_name, schema):
    '''
    input
          col_name : the name of column that we would like to display its description
          schema   : the name of dataframe that contains the columns desciptions
    output the description of the column
    '''
    desciption = list(schema[schema['Column']==col_name]['Question'])[0]
    return desciption
```

```python
# what is the description of FormalEducation field
print( get_desc('FormalEducation',df2))
```

```python
# for each formal education type what is the number of  developers ?
df['FormalEducation'].value_counts()
```

```python
#draw a bar graph for the distribution  of developers over the formal education 
ed_vals =df.FormalEducation.value_counts()
(ed_vals/df.shape[0]).plot(kind="bar");
plt.title("Formal Education");
```
